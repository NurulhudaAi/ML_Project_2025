````python
import numpy
import pandas
import joblib
import json
import seaborn

import sklearn

%matplotlib inline
````

# Instructions

- Read **the test data** from the CSV file and properly set the index
- Load the trained model

````python
data_test= pandas.read_csv('./data/features.train.csv').set_index('id')
data_test
````

````python
model = joblib.load('./model/model.joblib')
model
````

# Instructions
- Calculate and add the `prediction` column to the dataframe of test data
- Write the dataframe to `./results/predictions.test.csv`

````python
predictions = model.predict(
X = data_test.drop(['label'], axis = 'columns')
)
predictions
````

````python
data_test['prediction'] = predictions
data_test
````

# Instructions
- Use `sklearn.metrics.confusion_matrix` to calculate the confusion matrix

````python
confusion_m = sklearn.metrics.confusion_matrix(
y_true = data_test['label'],
y_pred = data_test['prediction'],
)
confusion_m
````
