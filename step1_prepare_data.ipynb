```python
import numpy as np
import pandas as pd
import sklearn.model_selection
```

```python
data = pd.read_csv('../work/data/high_salary.csv')
data
```

```python
data_train, data_test = sklearn.model_selection.train_test_split(
    data,
    train_size   = 0.80,
    shuffle      = True ,
    random_state = 0
)
```

```python
data_train.to_csv('./data/high_salary_train.csv')
data_test.to_csv('./data/high_salary_test.csv')
```
